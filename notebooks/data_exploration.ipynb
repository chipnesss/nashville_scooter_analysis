```python
from shapely.geometry import Point
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import folium
from folium.plugins import MarkerCluster
from folium.plugins import FastMarkerCluster
%matplotlib inline
```

```python
may_trip = pd.read_csv('../data/may_trip.csv')
```

```python
zipcodes = gpd.read_file('../data/zipcodes.geojson')
```

```python
may_trip['geometry'] = may_trip.apply(lambda x: Point((float(x.startLongitude),
                                                    float(x.startLatitude))),
                                   axis=1)
may_trip.head()
```

```python
zipcodes.crs
```

```python
may_trip_geo = gpd.GeoDataFrame(may_trip,
                               crs = zipcodes.crs,
                               geometry = may_trip['geometry'])
```

```python
type(may_trip_geo)
```

```python
zipcodes = zipcodes[['zip', 'po_name', 'geometry']]
```

```python
may_trip_byzip = gpd.sjoin(may_trip_geo, zipcodes, op = 'within')
```
